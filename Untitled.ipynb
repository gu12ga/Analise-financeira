```python
query = '''
query {
	tendency {
	  period
	  year
	  tendency
	  intercept
    company{
      numCnpj
      sigAgente
      scConjConsumidoras
    }
	}
}
'''
r = requests.post(url, json={'query': query}).json()
```

```python
data = []

for i in r['data']['tendency']:
    company = list(i.pop('company').values())
    tend = list(i.values())
    data.append(company+tend)
    
columns = ['numCnpj', 'sigAgente', 'scConjConsumidoras', 'period', 'year', 'tendency', 'intercept']


tendencies = pd.DataFrame(
    data,
    columns=columns
    
)
tendencies
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=2154)

print(X_train.shape)
print(y_train.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.score(X_test, y_test))
```

```python
svm = SVR()
svm.fit(X_train, y_train)
print(svm.score(X_test, y_test))
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
print(rf.score(X_test, y_test))
```

```python
trend_map = {}

for p, t in tendencies[['period', 'tendency']].values:
    if p not in trend_map:
        trend_map[int(p)] = t

trends = []
for i in decfec_df['numPeriodoIndice'].values:
    trends.append(trend_map.get(i, None))
```
